üì• 1. Chargement des donn√©es


```python
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv("/content/mm-5G.csv")

# Afficher les 5 premi√®res lignes
print(df.head())
```

üîç 2. Aper√ßu g√©n√©ral du dataset

```python
# Afficher les informations g√©n√©rales
print(df.info())

# Afficher les statistiques g√©n√©rales des variables num√©riques
print(df.describe())
```

Histogramme du d√©bit (Throughput) :

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df["Throughput"], bins=30, kde=True, color="royalblue")
plt.xlabel("D√©bit (Mbps)")
plt.ylabel("Fr√©quence")
plt.title("Distribution du D√©bit (Throughput)")
plt.show()
```

2Ô∏è‚É£ Boxplots pour d√©tecter les valeurs extr√™mes des indicateurs de signal

```python
signal_vars = ["lte_rsrp", "lte_rsrq", "nr_ssRsrp", "nr_ssRsrq", "lte_rssnr", "nr_ssSinr"]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[signal_vars])
plt.xticks(rotation=15)
plt.title("Boxplot des indicateurs de qualit√© du signal")
plt.show()
```

3Ô∏è‚É£ Heatmap (Carte de corr√©lation)

```python
# S√©lectionner uniquement les colonnes num√©riques
df_numeric = df.select_dtypes(include=['number'])

# G√©n√©rer la heatmap avec uniquement les variables num√©riques
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Carte de Corr√©lation entre les Variables")
plt.show()
```

4Ô∏è‚É£ Effet de la vitesse du mobile sur la connexion


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["movingSpeed"], y=df["Throughput"], alpha=0.7, color="darkgreen")
plt.xlabel("Vitesse du Mobile (km/h)")
plt.ylabel("D√©bit (Mbps)")
plt.title("Impact de la Vitesse du Mobile sur le D√©bit")
plt.show()
```
